```python
#Load all the data to dataFrames
import pandas as pd
import numpy as np
import re
import numpy as np
import matplotlib.pyplot as plt 
import matplotlib.dates as dates
import seaborn as sns
plt.style.use('fivethirtyeight')
#Create  dataframes
df_air_visit_data = pd.read_csv('.//air_visit_data.csv') #this is the training data
df_air_store_info = pd.read_csv('.//air_store_info.csv')
df_hpg_store_info = pd.read_csv('.//hpg_store_info.csv')
df_air_reserve = pd.read_csv('.//air_reserve.csv')
df_hpg_reserve = pd.read_csv('.//hpg_reserve.csv')
df_store_id_realation = pd.read_csv('.//store_id_relation.csv')
df_test_set = pd.read_csv('.//sample_submission.csv')
df_date_info = pd.read_csv('.//date_info.csv')
```

```python
df_air_reserve = pd.merge(df_air_reserve,df_air_store_info,on='air_store_id',how='inner')
```

```python
df_air_reserve.head()
```

```python
df_hpg_reserve = pd.merge(df_hpg_reserve,df_hpg_store_info,how='inner',on='hpg_store_id')
df_hpg_reserve.head()
```

```python
df_air_reserve_with_relation = pd.merge(df_air_reserve,df_store_id_realation,how='left',on='air_store_id')
df_air_reserve_with_relation.head()
```

```python
df_hpg_reserve_with_realtion = pd.merge(df_hpg_reserve,df_store_id_realation,how='left',on='hpg_store_id')
df_hpg_reserve_with_realtion.head()
```

```python
df_total_reserve = pd.merge(df_air_reserve_with_relation,df_hpg_reserve_with_realtion,how='outer')
df_total_reserve.head()
```

```python
df_total_reserve = df_total_reserve.drop(['hpg_store_id','hpg_genre_name','hpg_area_name'],axis=1)
df_total_reserve.head()
```

```python
len(df_total_reserve.air_store_id.unique())
```

```python
df_total_reserve = df_total_reserve.rename(columns={'visit_datetime':'visit_date'})
df_total_reserve['visit_date'] = pd.to_datetime(df_total_reserve['visit_date']).dt.date
df_air_visit_data['visit_date'] = pd.to_datetime(df_air_visit_data['visit_date']).dt.date
```

```python
df_total_visit_reserve = pd.merge(df_total_reserve,df_air_visit_data,how='right',on=['air_store_id','visit_date'])
```

```python
df_total_visit_reserve.head()
```

```python
#Adding new feature to get the time delay between visit date and reserved date
df_total_visit_reserve['time_delay'] = df_total_visit_reserve['visit_date'] - (pd.to_datetime(df_total_visit_reserve['reserve_datetime']).dt.date)
df_total_visit_reserve['time_delay'] = df_total_visit_reserve['time_delay'].dt.days
```

```python
#missing values are filled by 0's
df_total_visit_reserve['reserve_visitors'] = df_total_visit_reserve['reserve_visitors'].fillna(0)
df_total_visit_reserve['time_delay'] = df_total_visit_reserve['time_delay'].fillna(0)
```

```python
#Get all the store info
#Inner join df_hpg_store_info & df_store_id_realtion to get only required hpg stores which are mapped to air stores
df_hpg_store_info_joined = pd.merge(df_hpg_store_info,df_store_id_realation,how='inner',on='hpg_store_id')
df_hpg_store_info_joined = df_hpg_store_info_joined.drop('hpg_store_id',axis=1)
#Rename columns
df_hpg_store_info_joined = df_hpg_store_info_joined.rename(columns={'hpg_genre_name':'air_genre_name','hpg_area_name':'air_area_name'})
df_air_hpg_total_store_info = df_air_store_info.append(df_hpg_store_info_joined)
df_air_hpg_total_store_info.head()
```

```python
df_air_hpg_total_store_info.shape
```

```python
df_air_hpg_total_store_info = df_air_hpg_total_store_info.set_index('air_store_id')
```

```python
df_air_hpg_total_store_info.head()
```

```python
df_total_visit_reserve['latitude'] = df_total_visit_reserve['air_store_id'].apply(lambda x : df_air_hpg_total_store_info.loc[x]['latitude'] )
df_total_visit_reserve['longitude'] = df_total_visit_reserve['air_store_id'].apply(lambda x : df_air_hpg_total_store_info.loc[x]['longitude'])
df_total_visit_reserve['air_area_name'] = df_total_visit_reserve['air_store_id'].apply(lambda x : df_air_hpg_total_store_info.loc[x]['air_area_name'])
df_total_visit_reserve['air_genre_name'] = df_total_visit_reserve['air_store_id'].apply(lambda x : df_air_hpg_total_store_info.loc[x]['air_genre_name'])
```
